```python
path_to_annotations = "Annotations"
path_to_images = "..\JPEGImages"

dataset = importer.ImportVOC(path=path_to_annotations, path_to_images=path_to_images, name="VOC07")
dataset.df.head(20)
```

```python
# Note that the original VOC2007 train and test images are combined and there are now 10,000 images in total

print(f"Number of images: {dataset.analyze.num_images}")
print(f"Number of classes: {dataset.analyze.num_classes}")
print(f"Classes:{dataset.analyze.classes}")
print(f"Class counts:\n{dataset.analyze.class_counts}")
```

```python
# Split the dataset again into train, validation and test

dataset.splitter.GroupShuffleSplit(train_pct=.6, val_pct=.2, test_pct=.2)
dataset.analyze.ShowClassSplits()
```

```python
# Converted dataset in VOC07_YOLO
dataset.export.ExportToYoloV5(output_path=".\VOC07_YOLO\\labels", copy_images=True, yaml_file='VOC07_YOLO.yaml', use_splits=True)
```
